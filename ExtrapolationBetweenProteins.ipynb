In this notebook I am trying to test if a model improves its prediction performance when given fitness data from other proteins.

```python
import numpy as np
import os
import pickle as pkl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Read results from file
results_oneout = pkl.load(open("results_oneout.pkl", "rb"))
results_single = pkl.load(open("results_single.pkl", "rb"))
results_mix = pkl.load(open("results_mix.pkl", "rb"))

# Bar plot comparing results_oneout, results_single and results_mix
fig, ax = plt.subplots()
width = 0.25
x = np.arange(len(results_oneout))
ax.bar(x - width/2, results_oneout.values(), width, label="One out")
ax.bar(x + width/2, results_single.values(), width, label="Single")
ax.bar(x + 3*width/2, results_mix.values(), width, label="Mix")
ax.set_ylabel("MSE")
ax.set_xticks(x)
ax.set_xticklabels(results_oneout.keys())
ax.legend()
plt.show()
```
